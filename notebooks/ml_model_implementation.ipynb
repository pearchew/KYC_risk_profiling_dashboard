```python
import joblib
import pandas as pd
import numpy as np

# 1. Load the "brains" back into memory
loaded_model = joblib.load('../models/best_fraud_model.joblib')
loaded_scaler = joblib.load('../models/robust_scaler.joblib')

# 2. Imagine a new credit card swipe comes in from a web app or database
# (This simulates a new transaction with fake feature values)
transactions = pd.read_csv('../data/credit_card_user_data.csv')
transactions_dropped = transactions.drop(['Customer_ID', 'Class'], axis=1)

# 3. Preprocess the new data EXACTLY how you did in the notebook
transactions_dropped['scaled_amount'] = loaded_scaler.transform(transactions_dropped['Amount'].values.reshape(-1,1))
transactions_dropped['scaled_time'] = loaded_scaler.transform(transactions_dropped['Time'].values.reshape(-1,1))

# Drop the original columns and reorder to match your training data
transactions_dropped.drop(['Time', 'Amount'], axis=1, inplace=True)
# Note: Ensure the columns are in the exact same order as X_train!

# 4. Make the instant prediction!
prediction = loaded_model.predict(transactions_dropped)

results_df = pd.DataFrame({
    'Customer_ID': transactions['Customer_ID'],
    'Is_Fraud': prediction,
    'Time': transactions['Time'],
    'Amount': transactions['Amount']
})
```
