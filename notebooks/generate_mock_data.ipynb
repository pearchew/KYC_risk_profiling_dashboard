```python
import pandas as pd
import random
from faker import Faker
import os

# Initialize Faker and set seeds for reproducibility
fake = Faker()
Faker.seed(42)
random.seed(42)

# Define pools of data, intentionally including high-risk scenarios
nationalities = ['USA', 'UK', 'Canada', 'Panama', 'Cayman Islands', 'North Korea', 'Singapore', 'Australia']
occupations = ['Software Engineer', 'Teacher', 'Casino Operator', 'Accountant', 'Cash-Intensive Business Owner', 'Doctor', 'Freelancer']

# Generate 100 customer records
mock_data = []
for _ in range(100):
    customer = {
        # .unique ensures it never generates the exact same ID twice.
        # bothify tells it to replace specific symbols: # gets replaced by a random number (0-9), and ? gets replaced by a random letter (A-Z).
        # So, CUST-####-???? turns into something like CUST-4921-XPLQ.
        'Customer_ID': fake.unique.bothify(text='CUST-####-????'), # e.g., CUST-1234-ABCD
        'Age': random.randint(18, 85),
        'Nationality': random.choice(nationalities),
        'Occupation': random.choice(occupations),
        'Estimated_Annual_Income': random.randint(20000, 500000)
    }
    mock_data.append(customer)

# Convert to a Pandas DataFrame
df_kyc = pd.DataFrame(mock_data)

# Ensure the data directory exists (just in case)
os.makedirs('../data', exist_ok=True)

# Save the dataset to the 'data' folder
csv_path = '../data/mock_kyc_data.csv'
df_kyc.to_csv(csv_path, index=False)

print(f"Successfully generated {len(df_kyc)} records and saved to {csv_path}")

# Preview the first 5 rows
df_kyc.head()
```
